**Conclusion:**
The Indian agriculture dataset (1966–2017) clearly shows that rice and wheat dominated in terms of both cultivated area and production, with Punjab, Uttar Pradesh, and West Bengal consistently performing as top contributors. Post-Green Revolution, yields of these crops improved significantly (from 800–1,400 kg/ha to over 2,000–5,000+ kg/ha in many districts). Millets and pulses were mainly grown in rainfed regions like Rajasthan, Madhya Pradesh, and Maharashtra. Sugarcane and vegetables also showed strong growth, especially in Maharashtra and West Bengal. However, eastern states still lag in yield compared to northern irrigated regions. This highlights the urgent need for better irrigation, crop diversification, and sustainable practices to ensure long-term food security in the face of climate change and resource constraints.
